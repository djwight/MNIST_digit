# TF model prediction MNIST digits

Notebook conayins the code for prediction using the base_model and base_plus_model trained on the AWS GPU.

```python
from pathlib import Path
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import tensorflow
from keras.models import save_model, load_model
```

```python
Path.cwd()
```

```python
#Set up dirs for project
base_dir= Path.cwd().parents[0]
raw_data_dir= base_dir/'data'/'raw'
interim_data_dir= base_dir/'data'/'interim'
model_dir= base_dir/'models'
```

```python
test_data= pd.read_csv(raw_data_dir/'test.csv')
test_data.head()
```

```python
test_data_std= test_data/254
test_data_std.describe()
```

```python
test_data_vals= test_data_std.values
```

```python
os.chdir(model_dir)
Path.cwd()
```

```python
base_plus_model= load_model('base_plus_model.h5')
```

```python
preds= base_plus_model.predict_classes(test_data_vals)
```

```python
pred_data= pd.DataFrame({'ImageId': range(1, 28001)})
```

```python
pred_data['label']= preds
```

```python
pred_data.head()
```

```python
pred_data.to_csv(interim_data_dir/'base_plus_pred.csv', index=False)
```
